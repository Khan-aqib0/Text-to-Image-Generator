```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "stabilityai/stable-diffusion-2"
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    safety_checker=None
)
pipe.to("cuda")
```

```python
prompt = "A man named Cyfuture, dressed sharply, confidently standing in front of a modern corporate building with the logo Cyfuture displayed above the entrance"
image = pipe(prompt).images[0]
image.save("generated_image.png")
```
